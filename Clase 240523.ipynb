```python
import pandas as pd 
agosto2014=pd.read_excel("VentasGroucery_Meses (1).xlsx", sheet_name="Ventas Agosto 2014") 
```

```python
agosto2014=pd.read_excel("VentasGroucery_Meses (1).xlsx", sheet_name="Ventas Agosto 2014") agosto2014
```

```python
septiembre2014=pd.read_excel("VentasGroucery_Meses (1).xlsx", sheet_name="Ventas Septiembre 2014")
septiembre2014
```

```python
#Como unir 2 o + DataFrames que tengan la misma estructura
#es decir que tenga el mismo numero de campos o columnas y con los 
#mismos nombres, usaremos concat
dosmeses=pd.concat([agosto2014,septiembre2014])
dosmeses
```

```python
#vamos a crear un campo nuevo llamado Amount=Unit Price x Quantity
dosmeses["Amount"]=dosmeses["Unit Price"]*dosmeses["Quantity"]
dosmeses
```

```python
#vamos a crear una agrupacion para ver la suma de amount por categoria
dosmeses.groupby(["Category"]).Amount.sum()
```

```python
#Convertir el resultado de un group by, que genera una serie 
#en un data frame 
dosmesesgroup=dosmeses.groupby(["Category"]).Amount.sum()
dosmesesDgroup = pd.DataFrame(dosmesesgroup, columns=["Amount"])
dosmesesDgroup
```

```python
dosmesesDgroup.info()
```

```python
#Al convertirlo en DataFrame la categoria la considera
#un indice y no un campo
#converitr el indice en un campo con reset_index
dosmesesDFgroup = dosmesesDgroup.reset_index()
dosmesesDFgroup
```

```python
import matplotlib.pyplot as plt
```

```python
#Los graficos en Python utilizando la libreria Matplotlib
#pip install matplotlib en cmd
#crear la figura
fig, ax=plt.subplots() 

#crearemos un grafico circular o de quesitos con pie
ax.pie([8,2,5,3,6])
#orden de imprimir 
plt.show();
```

```python
#vamos a crear el grafico de quesitos del DataFrame dos meses Dfgroup
#que tiene la suma de Amount por categoria 
import matplotlib.pyplot as plt
fig, ax=plt.subplots()
ax.pie(dosmesesDFgroup["Amount"])
plt.show();
```

```python
fig, ax=plt.subplots()
ax.pie(dosmesesDFgroup["Amount"],labels=dosmesesDFgroup["Category"])
plt.show();
```

```python
#poner el porcentaje en el quesito con el autopct
fig, ax=plt.subplots()
ax.pie(dosmesesDFgroup["Amount"],labels=dosmesesDFgroup["Category"],
      autopct="%1.f%%")
plt.show();
```
